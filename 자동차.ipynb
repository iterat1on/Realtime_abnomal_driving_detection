## 시간 -> 분 변환

```python
def calc_time(x) :
  x = x.replace('분', '')
  if '시간' in x :
    try :
      hour, min = x.split('시간 ')
      a = int(hour)*60 + int(min)
    except ValueError:
      hour = x.replace('시간', '')
      a = int(hour)*60
  else :
    a = int(x)
  return a
```

```python
df['due_min'] = df.duration.apply(lambda x : calc_time(x))
```

```python
df['due_min'].value_counts()
```

## 요금 변환

```python
df.fee.replace(',','',regex=True,inplace=True)
df.fee.replace('원','',regex=True,inplace=True)
df.fee=df.fee.apply(pd.to_numeric)
```

## 거리 -> KM 변환

```python
df.reset_index(inplace= True)
df.drop('index', axis =1, inplace = True)

def root_dist_def(x) :

  if 'km' in x :
    pass
  else :
    x = x.replace('m', '')
    x = float(x)
    x = x * 0.001
    x = str(x)
    x = x + 'km'
    x = x.replace('km', '')
    x = float(x)
  return x

df['root_dist'] = df['root_dist'].apply(root_dist_def)
df.rename(columns={'root_dist': 'root_dist(km)'}, inplace = True)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

## 원핫인코딩

```python
df_result = pd.DataFrame(te_result, columns=te.columns_)
```

### 최소지지도 0.01
