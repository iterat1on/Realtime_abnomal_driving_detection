```python
df.columns = ['region','total','total_m','total_w','0g_pop','m_0g_pop','w_0g_pop','20g1_pop','m_20g1_pop','w_20g1_pop','20g2_pop','m_20g2_pop','w_20g2_pop','30g1_pop','m_30g1_pop','w_30g1_pop','30g2_pop','m_30g2_pop','w_30g2_pop','40g1_pop','m_40g1_pop','w_40g1_pop','40g2_pop','m_40g2_pop','w_40g2_pop', '50g1_pop','m_50g1_pop','w_50g1_pop','50g2_pop','m_50g2_pop','w_50g2_pop', '60g1_pop','m_60g1_pop','w_60g1_pop','60g2_pop','m_60g2_pop','w_60g2_pop','70g1_pop','m_70g1_pop','w_70g1_pop','70g2_pop','m_70g2_pop','w_70g2_pop','80g1_pop','m_80g1_pop','w_80g1_pop','80g2_pop','m_80g2_pop','w_80g2_pop']
df.columns
```

```python
pop = df
pop = pop.drop(['0g_pop','m_0g_pop','w_0g_pop','20g1_pop','m_20g1_pop','w_20g1_pop','20g2_pop','m_20g2_pop','w_20g2_pop','30g1_pop','m_30g1_pop','w_30g1_pop','30g2_pop','m_30g2_pop','w_30g2_pop','40g1_pop','m_40g1_pop','w_40g1_pop','40g2_pop','m_40g2_pop','w_40g2_pop', '50g1_pop','m_50g1_pop','w_50g1_pop','50g2_pop','m_50g2_pop','w_50g2_pop', '60g1_pop','m_60g1_pop','w_60g1_pop','60g2_pop','m_60g2_pop','w_60g2_pop','70g1_pop','m_70g1_pop','w_70g1_pop','70g2_pop','m_70g2_pop','w_70g2_pop','80g1_pop','m_80g1_pop','w_80g1_pop','80g2_pop','m_80g2_pop','w_80g2_pop'], axis=1)
```

```python
pop1 = df[['m_0g_pop','w_0g_pop']]
pop2 = df[['m_20g1_pop','w_20g1_pop','m_20g2_pop','w_20g2_pop']]
pop3 = df[['m_30g1_pop','w_30g1_pop','m_30g2_pop','w_30g2_pop']]
pop4 = df[['m_40g1_pop','w_40g1_pop','m_40g2_pop','w_40g2_pop']]
pop5 = df[['m_50g1_pop','w_50g1_pop','m_50g2_pop','w_50g2_pop']]
pop6 = df[['m_60g1_pop','w_60g1_pop','m_60g2_pop','w_60g2_pop']]
pop7 = df[['m_70g1_pop','w_70g1_pop','m_70g2_pop','w_70g2_pop']]
pop8 = df[['m_80g1_pop','w_80g1_pop','m_80g2_pop','w_80g2_pop']]
```

```python
pop['0g']=pop1.sum(axis=1)
pop['20g']=pop2.sum(axis=1)
pop['30g']=pop3.sum(axis=1)
pop['40g']=pop4.sum(axis=1)
pop['50g']=pop5.sum(axis=1)
pop['60g']=pop6.sum(axis=1)
pop['70g']=pop7.sum(axis=1)
pop['80g']=pop8.sum(axis=1)
```

```python
pop.drop(0, inplace=True)
pop
```

```python
gr1 = pop[['0g', '20g', '30g']] #청년층
gr2 = pop[['40g', '50g']] #중년층
gr3 = pop[['60g', '70g', '80g']] #장년층
pop['gr1'] = gr1.sum(axis=1)
pop['gr2'] = gr2.sum(axis=1)
pop['gr3'] = gr3.sum(axis=1)
pop
```

```python
import requests
import json

r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
m = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=11, 
)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(m)

m
```
