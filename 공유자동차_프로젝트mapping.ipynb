### 필요한 변수만 추출

### gps 좌표 한 번에 입력을 위한 위/경도 좌표 통합 columns 생성

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
df["start_gps"] = df["s_lat"].map(str) + ", " + df["s_lng"].map(str)
df["end_gps"] = df["e_lat"].map(str) + ", " + df["e_lng"].map(str)
df.head()
```
