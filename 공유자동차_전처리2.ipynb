```python
import pandas as pd
import numpy as np
import math
import pandas as pd
```

```python
df = pd.read_csv('./drive/MyDrive/airport_filtered.csv')
df.head()
```

#전처리

##road_split

```python
df['road_1'] = df["road_list"].str.split(" ", expand=True)[1]
df['road_2'] = df["road_list"].str.split(" ", expand=True)[4]
df['road_3'] = df["road_list"].str.split(" ", expand=True)[7]
df.tail()
```

```python
df['road_1'].fillna('-',inplace=True)
df['road_2'].fillna('-',inplace=True)
df['road_3'].fillna('-',inplace=True)
```

```python
df['path'] = df['road_1'] +',' + df['road_2'] +',' +  df['road_3']
```

```python
df.path.replace('-','',regex=True,inplace=True)
for index in range(0,len(df.path)):
   df.path[index] = df.path[index].strip(',')
```

```python
df
```

##같은 주소 제거 

```python
df.dropna(subset = ['road_list'], inplace = True)
df.reset_index(inplace= True)
#df.drop('index', axis =1, inplace = True)
df.info()
```

## 시간(duration) 전처리

```python
def calc_time(x) :
  x = x.replace('분', '')
  if '시간' in x :
    try :
      hour, min = x.split('시간 ')
      a = int(hour)*60 + int(min)
    except ValueError:
      hour = x.replace('시간', '')
      a = int(hour)*60
  else :
    a = int(x)
  return a
```

```python
df['duration(min)'] = df.duration.apply(lambda x : calc_time(x))
df.drop('duration', axis =1, inplace = True)
df.head()
```

##요금(fee) 전처리

```python
df.fee.replace(',','',regex=True,inplace=True)
df.fee.replace('원','',regex=True,inplace=True)
df.fee = df.fee.apply(pd.to_numeric)
```

##총 거리(root_dist) 전처리

```python
def root_dist_def(x) :
  if 'km' in x :
    x = x.replace('km', '')
    pass
  else :
    x = float(x.replace('m', ''))*0.001
  return x
```

```python
df['root_dist(km)'] = df.root_dist.apply(lambda x : root_dist_def(x))
df.drop('root_dist', axis =1, inplace = True)
df.head()
```

##시간대 변수 추가

```python
#datetime으로 변환
df['s_Datetime'] = pd.to_datetime(df['s_date'], format="%Y-%m-%d %H:%M")
df['e_Datetime'] = pd.to_datetime(df['e_date'], format="%Y-%m-%d %H:%M")

#시작날짜 분리
df['s_year'] = df['s_Datetime'].dt.year
df['s_month'] = df['s_Datetime'].dt.month
df['s_day'] = df['s_Datetime'].dt.day
df['s_hour'] = df['s_Datetime'].dt.hour


#종료날짜 분리
df['e_year'] = df['e_Datetime'].dt.year
df['e_month'] = df['e_Datetime'].dt.month
df['e_day'] = df['e_Datetime'].dt.day
df['e_hour'] = df['e_Datetime'].dt.hour

#요일 추출
df['s_week'] = df['s_Datetime'].dt.weekday
df['e_week'] = df['e_Datetime'].dt.weekday

#열 제거
df.drop('s_date', axis =1, inplace = True)
df.drop('e_date', axis =1, inplace = True)
```

```python
def weeks(x):

  if x <= 4:
    return '평일'
  
  else : 
    return '주말'
```

```python
def week(x):

  if x == 0:
    return '월'
  elif x == 1:
    return '화'
  elif x == 2:
    return '수'
  elif x == 3:
    return '목'
  elif x == 4:
    return '금'
  elif x == 5:
    return '토'
  else : 
    return '일'
```

```python
#요일을 평일,주말로 변환
df['s_weeks'] = df['s_week'].apply(lambda x: weeks(x))
df['e_weeks'] = df['e_week'].apply(lambda x: weeks(x))
```

```python
#요일을 한글로 변환
df['s_week'] = df['s_week'].apply(lambda x: week(x))
df['e_week'] = df['e_week'].apply(lambda x: week(x))
df.head()
```

```python
df.s_week.value_counts()
```

```python
#출발일과 도착일이 다른 데이터(년도와 달은 모두 같음)
df[df['s_day'] != df['e_day']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
te = TransactionEncoder()
te_result = te.fit(transaction).transform(transaction)
te_result
```

##원핫인코딩

```python
df_2 = pd.DataFrame(te_result, columns=te.columns_)
```

##최소지지도 : 0.01

```python
frequent_itemsets = apriori(df_2, min_support=0.01, use_colnames=True)
```

```python
frequent_itemsets.sort_values('support', ascending= False)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(df_filtered)
scaled_df = pd.DataFrame(scaled)
scaled_df.head()
```

```python
scaled_df.corr()
```

##PCA

```python
from sklearn.decomposition import PCA
X = scaled_df.copy()

pca = PCA(n_components=2)

pca.fit(X)
x_pca = pca.transform(X)
x_pca
```

```python
pca_df = pd.DataFrame(x_pca)
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
clusters = kmeans.fit(pca_df)
pca_df['clusters'] = clusters.labels_
```

```python
pca_df.clusters.value_counts()
```
